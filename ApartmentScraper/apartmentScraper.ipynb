```python
import os.path
from os import path
import pathlib 


import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

import pandas as pd 
```

```python
#Checking if file exists. 
column_names = ['Link', 'Adresse', 'Prisantydning', 'Fellesgjeld', 'Totalpris', 'Felleskostnader', 'Areal', 'Eierform', 'Soverom']


#file = pathlib.Path('./boligjakten.csv')
file = pathlib.Path('./boligjakten.csv')

if file.exists():
    print('You have a file. Loading file...')
    clean_dataset = pd.read_csv('./boligjakten.csv')
    print('File loaded!')
else:
    print('No file found. Writing new...')
    clean_dataset = pd.DataFrame(data=None, columns=column_names)
    clean_dataset.to_csv('./boligjakten.csv')
    print('File created!')
    
    
#Scraping data
url = input('LINK: ')

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url)

prisantydning = ((driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/div[2]/span[2]').text).replace('kr', '')).replace(' ','')
fellesgjeld = ((driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/div[2]/dl[2]/dd[1]').text).replace('kr', '')).replace(' ','')
totalpris = ((driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/div[2]/dl[2]/dd[3]').text).replace('kr', '')).replace(' ','')
felleskost = ((driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/div[2]/dl[2]/dd[4]').text).replace('kr', '')).replace(' ','')
areal = ((driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/section[2]/dl/dd[5]').text).replace('mÂ²', '')).replace(' ','')
eierform = driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/section[2]/dl/dd[2]').text
soverom = driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/section[2]/dl/dd[3]').text
adresse = driver.find_element(by=By.XPATH, value='//*[@id="realestateClassifiedContainer"]/div[2]/div/div[1]/div/section[5]/p').text



print('Prisantydning:     ',prisantydning)
print('Fellesgjeld:       ',fellesgjeld)
print('Totalpris:         ',totalpris)
print('Felleskostnader:   ',felleskost)
print('Areal :            ',areal)
print('Eierform:          ',eierform)
print('Soverom:           ',soverom)
print('Adresse:           ',adresse)
#print('Link:              ',url)


#printing it as a dataframe to easily transfer
datas = ({'Link':[url], 
         'Adresse':[adresse],
         'Prisantydning':[prisantydning],
         'Fellesgjeld': [fellesgjeld],
         'Totalpris':[totalpris], 
         'Felleskostnader':[felleskost],
         'Areal':[areal],
         'Eierform':[eierform],
         'Soverom': [soverom]})
#datas = {[url],[adresse],[prisantydning],[fellesgjeld],[totalpris],[felleskost],[areal],[eierform],[soverom]}

new_dataset = pd.DataFrame(data=datas)
saving_dataset = clean_dataset.append(new_dataset, ignore_index = True)


saving_dataset.to_csv('./boligjakten.csv', index=False)

saving_dataset
```
